```python
df_player_attributes_clean['date'] = pd.to_datetime(df_player_attributes_clean['date'])

df_player_attributes_clean['overall_rating'].fillna(df_player_attributes_clean['overall_rating'].mean(), inplace=True)
df_player_attributes_clean['potential'].fillna(df_player_attributes_clean['potential'].mean(), inplace=True)
```

```python
d  = (df_player_attributes_clean.isnull().sum() / len(df_player_attributes_clean) *100)
d
```

```python
# Remplacement par la moyenne pour les colonnes numériques
columns_to_impute_mean = ['stamina', 'strength', 'vision']
for col in columns_to_impute_mean:
    df_player_attributes_clean[col].fillna(df_player_attributes_clean[col].mean(), inplace=True)
```

```python
columns_to_impute_mode = ['attacking_work_rate', 'defensive_work_rate']
for col in columns_to_impute_mode:
    df_player_attributes_clean[col].fillna(df_player_attributes_clean[col].mode()[0], inplace=True)
```

```python
d  = (df_player_attributes_clean.isnull().sum() / len(df_player_attributes_clean) *100)
d
```
